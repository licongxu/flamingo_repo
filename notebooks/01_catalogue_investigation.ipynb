# 01 - Catalogue headers

First look at the FLAMINGO **hydro L2p8_m9** SOAP halo catalogue under `data/hydro_L2p8m9/catalogue/`. Here we just load it through the `flamingo` package and print the column headers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Publication-quality plot defaults
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "serif",
    "font.size": 11,
    "axes.labelsize": 12,
    "axes.titlesize": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "figure.dpi": 100,
    "savefig.dpi": 300,
    "axes.linewidth": 0.8,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.top": True,
    "ytick.right": True,
})

from flamingo.catalogue import efunc, angular_diameter_distance
from flamingo import paths
from flamingo.catalogue import load_catalogue

print('catalogue:', paths.HYDRO_CATALOGUE)
```

```python
df = load_catalogue(paths.HYDRO_CATALOGUE)
print(f'rows = {len(df):,}   columns = {df.shape[1]}')
```

## Column headers

```python
for i, col in enumerate(df.columns):
    print(f'{i:2d}  {col}')
```

### $Y-M$ Scaling Relations

```python
z = df['z'].to_numpy(float)
M500 = df['M_500c_Msun'].to_numpy(float)
Y5R500 = df['Y_5R500c_Mpc2'].to_numpy(float)

Ez = efunc(z)
dA = angular_diameter_distance(z)              # proper Mpc (hmfast, physical)
Y_scaled = Ez ** (-2.0 / 3.0) * Y5R500
# Y_scaled = Ez ** (-2.0 / 3.0) * Y5R500 * dA ** 2

good = np.isfinite(Y_scaled) & (Y_scaled > 0) & (M500 > 0)
print(f'{good.sum():,} clusters with positive, finite scaled-Y')
```

```python
fig, ax = plt.subplots(figsize=(6.5, 5.2))
hb = ax.hexbin(M500[good], Y_scaled[good], xscale='log', yscale='log',
               gridsize=60, bins='log', mincnt=1, cmap='viridis')
cb = fig.colorbar(hb, ax=ax)
cb.set_label(r'$\log_{10} N$')

# median trend in log-mass bins
mbins = np.logspace(np.log10(M500[good].min()), np.log10(M500[good].max()), 18)
mc = 0.5 * (mbins[:-1] + mbins[1:])
idx = np.digitize(M500[good], mbins) - 1
med = np.array([np.median(Y_scaled[good][idx == b]) if (idx == b).any() else np.nan
                for b in range(len(mc))])
# ax.plot(mc, med, 'o-', color='crimson', ms=4, lw=1.5, label='median')

ax.set_xlabel(r'$M_{500c}\ [M_\odot]$')
ax.set_ylabel(r'$E(z)^{-2/3}\,Y_{500c}\,d_A^2\ [\mathrm{Mpc^4}]$')
ax.set_title(r'FLAMINGO L2p8\_m9 hydro: tSZ flux-mass relation')
ax.legend()
fig.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# Prepare log-log data for regression
logM = np.log10(M500[good])
logY = np.log10(Y_scaled[good])

# Perform linear regression in log-log space
slope, intercept, r_value, p_value, std_err = stats.linregress(logM, logY)
amplitude = 10**intercept

print(f"Best-fit log-log regression:")
print(f"    log10(Y) = {intercept:.4f} + {slope:.4f} * log10(M)")
print(f"    Y = {amplitude:.4e} * M^{slope:.4f}")

# Plotting
fig, ax = plt.subplots(figsize=(6.5, 5.2))
hb = ax.hexbin(M500[good], Y_scaled[good], xscale='log', yscale='log',
               gridsize=60, bins='log', mincnt=1, cmap='viridis')
cb = fig.colorbar(hb, ax=ax)
cb.set_label(r'$\log_{10} N$')

# Overplot regression line
Mfit = np.logspace(np.log10(M500[good].min()), np.log10(M500[good].max()), 100)
Yfit = amplitude * Mfit**slope

# Use a raw string for the label, double backslashes only where truly needed,
# and be careful not to create invalid escape sequences.
label_str = r'Best-fit: $Y = %.2e\,M^{%.2f}$' % (amplitude, slope)
ax.plot(Mfit, Yfit, color='crimson', lw=2.5, label=label_str)

ax.set_xlabel(r'$M_{500c}\ [M_\odot]$')
ax.set_ylabel(r'$E(z)^{-2/3}\,Y_{500c}\,d_A^2\ [\mathrm{Mpc^2}]$')
ax.set_title(r'FLAMINGO L2p8\_m9 hydro: tSZ flux-mass relation')

# Only add legend if there are labeled artists
handles, labels = ax.get_legend_handles_labels()
if len(labels) > 0:
    ax.legend()

fig.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# ========= 1. Top 2000 most massive clusters =========

# Select indices of the 2000 most massive clusters (among those marked as "good")
top2000_massive_idx = np.argsort(M500[good])[-2000:]

# Subset the arrays to only these 2000 most massive clusters
M500_top2000 = M500[good][top2000_massive_idx]
Y_scaled_top2000 = Y_scaled[good][top2000_massive_idx]

# Prepare log-log data for regression
logM_massive = np.log10(M500_top2000)
logY_massive = np.log10(Y_scaled_top2000)

# Perform linear regression in log-log space
slope_massive, intercept_massive, r_value, p_value, std_err = stats.linregress(logM_massive, logY_massive)
amplitude_massive = 10**intercept_massive

print(f"Best-fit log-log regression (2000 most massive clusters):")
print(f"    log10(Y) = {intercept_massive:.4f} + {slope_massive:.4f} * log10(M)")
print(f"    Y = {amplitude_massive:.4e} * M^{slope_massive:.4f}")

# Plotting
fig, ax = plt.subplots(figsize=(6.5, 5.2))
hb = ax.hexbin(M500_top2000, Y_scaled_top2000, xscale='log', yscale='log',
               gridsize=60, bins='log', mincnt=1, cmap='viridis')
cb = fig.colorbar(hb, ax=ax)
cb.set_label(r'$\log_{10} N$')

# Overplot regression line
Mfit_massive = np.logspace(np.log10(M500_top2000.min()), np.log10(M500_top2000.max()), 100)
Yfit_massive = amplitude_massive * Mfit_massive**slope_massive

label_str_massive = r'Best-fit: $Y = %.2e\,M^{%.2f}$' % (amplitude_massive, slope_massive)
ax.plot(Mfit_massive, Yfit_massive, color='crimson', lw=2.5, label=label_str_massive)

ax.set_xlabel(r'$M_{500c}\ [M_\odot]$')
ax.set_ylabel(r'$E(z)^{-2/3}\,Y_{500c}\,d_A^2\ [\mathrm{Mpc^2}]$')
ax.set_title(r'FLAMINGO L2p8\_m9 hydro: tSZ flux-mass relation (2000 most massive clusters)')

handles, labels = ax.get_legend_handles_labels()
if len(labels) > 0:
    ax.legend()

fig.tight_layout()
plt.show()
```

```python
# ========= 2. Top 2000 lowest-redshift clusters =========

# Select indices of the 2000 clusters with lowest z (among those marked as "good")
top2000_lowz_idx = np.argsort(z[good])[:2000]

# Subset the arrays to only these 2000 clusters
M500_lowz2000 = M500[good][top2000_lowz_idx]
Y_scaled_lowz2000 = Y_scaled[good][top2000_lowz_idx]

# Prepare log-log data for regression
logM_lowz = np.log10(M500_lowz2000)
logY_lowz = np.log10(Y_scaled_lowz2000)

# Perform linear regression in log-log space
slope_lowz, intercept_lowz, r_value, p_value, std_err = stats.linregress(logM_lowz, logY_lowz)
amplitude_lowz = 10**intercept_lowz

print(f"Best-fit log-log regression (2000 lowest-z clusters):")
print(f"    log10(Y) = {intercept_lowz:.4f} + {slope_lowz:.4f} * log10(M)")
print(f"    Y = {amplitude_lowz:.4e} * M^{slope_lowz:.4f}")

# Plotting
fig, ax = plt.subplots(figsize=(6.5, 5.2))
hb = ax.hexbin(M500_lowz2000, Y_scaled_lowz2000, xscale='log', yscale='log',
               gridsize=60, bins='log', mincnt=1, cmap='viridis')
cb = fig.colorbar(hb, ax=ax)
cb.set_label(r'$\log_{10} N$')

# Overplot regression line
Mfit_lowz = np.logspace(np.log10(M500_lowz2000.min()), np.log10(M500_lowz2000.max()), 100)
Yfit_lowz = amplitude_lowz * Mfit_lowz**slope_lowz

label_str_lowz = r'Best-fit: $Y = %.2e\,M^{%.2f}$' % (amplitude_lowz, slope_lowz)
ax.plot(Mfit_lowz, Yfit_lowz, color='crimson', lw=2.5, label=label_str_lowz)

ax.set_xlabel(r'$M_{500c}\ [M_\odot]$')
ax.set_ylabel(r'$E(z)^{-2/3}\,Y_{500c}\,d_A^2\ [\mathrm{Mpc^2}]$')
ax.set_title(r'FLAMINGO L2p8\_m9 hydro: tSZ flux-mass relation (2000 lowest-$z$ clusters)')

handles, labels = ax.get_legend_handles_labels()
if len(labels) > 0:
    ax.legend()

fig.tight_layout()
plt.show()
```
